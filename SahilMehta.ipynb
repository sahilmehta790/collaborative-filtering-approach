```python
import pandas as pd
from scipy import spatial
import numpy as np 
```

```python
df = pd.read_csv('data_recsys.csv')
df_new =  pd.read_csv('data_recsys.csv',index_col=0)
```

```python
df
```

```python
df['avg'] = df.iloc[:, 1:8].mean(axis=1)
```

```python
list_col=[]
for i in df.columns:
    if i == 'Unnamed: 0':
        continue
    elif i == 'avg':
        continue
    else:
        list_col.append(i)
```

```python
list_col
```

```python
for j in list_col:
    df[j] = df[j]-df['avg']
```

```python
df
```

```python
df.set_index('Unnamed: 0',inplace = True)
```

```python
df
```

```python
focal_user, target_artist = "Angelica", "Drake"
```

```python
del df['avg']
```

```python
for index,row in df.iterrows():
    print(index)
```

```python
col_1 = ~df.loc[index].isna()
```

```python
dict1 = {}
cosine_similarity = lambda U1,U2: np.dot(U1,U2)/(np.linalg.norm(U1)*np.linalg.norm(U2)) #lambda function to calculate cosine similarity
for index,row in df.iterrows():
    if index == focal_user:
        continue
    else:
        cola = ~df.loc[index].isna()#ignore if the value is NA
        colb = ~df.loc[focal_user].isna()#ignore if the value is NA
        
        U1 = df.loc[index][(cola*colb).values]
        U2 = df.loc[focal_user][(cola*colb).values]
        dict1[index] = cosine_similarity(U1,U2)
 
```

```python
dict1
```

```python
## calculate the predicted score 
req_score = df[target_artist][~df[target_artist].isna()]#getting the taget artist score for other users
focal_ua = df_new.mean(axis=1)[focal_user] #focal user average score

nume = 0
dino = 0

for i,x in req_score.to_frame().iterrows(): #as its a series object
    if i == focal_user:
        continue
    else:
        nume += dict1[i]*req_score[i]
        dino += abs(dict1[i])

score = focal_ua + (nume/dino)  
print(score)
```
